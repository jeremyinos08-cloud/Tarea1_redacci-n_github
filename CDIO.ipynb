## Neumáticos (Roce)
Para entender la dinámica del vehículo, primero debemos analizar las fuerzas y momentos que se generan en el neumático en las tres direcciones del espacio.
Los neumáticos son el único punto de contacto entre el vehículo y el suelo. Cada neumático transmite fuerzas a través de una superficie pequeña llamada huella de contacto.

Dirección z: Soporte y amortiguación.

Dirección x: Tracción y frenado.

Dirección y: El guiado y las curvas.

![Neumático](Figs/neumatico.png)

```python
Fx = f"{x} (N)"
Fy = f"{x} (N)"
Fz = f"{x} (N)"
mu = x # mientras mas blando el material, mayor es el coeficiente de fricción"

# Process
F_r = mu * Fz
F_total = (Fx**2 + Fy**2)**0.5

# Output
print(f"Friction force: {F_r}")
print(f"Total force: {F_total}")
```

## Neumáticos (Deriva)

Es casi un deslizamiento, no un derrape (deslizamiento mecánico puro), sino una deformación elástica continua del caucho en la huella de contacto.

El ángulo de la deriva es la diferencia angular entre la dirección hacia donde apunta geométricamente la rueda y la dirección real en la que se está moviendo el centro de la rueda.

![angle](Figs/angle.png)

```python
# Input
import math

Vx = x
"(m/s)"  # velocidad longitudinal de avance
Vy = x
"(m/s)"  # velocidad lateral de la rueda
k_alpha = x
"(N/rad)"  # constante de deslizamiento

# Process
alpha = math.atan(Vy / Vx)  # Ángulo de deslizamiento
Y = k_alpha * alpha  # Fuerza de deslizamiento

# Output
print(f"Slip angle: {math.degrees(alpha)} degrees")
print(f"Sliding force: {Y}")
```

## 3. Resistencia aerodinámica

```python
# Input
V_fluido = x "(m/s)"
Rho_fluido = x "(kg/m^3)}"
A_front = x "(m^2)"
C_d = x "(dimensionless)"

# Process
F_drag = 0.5 * Rho_fluido * V_fluido^2 * A_front * C_d

# Output
print(f"Drag force: {F_drag} N")
```

## 4. Potencia mecánica

```python
# Input
#F_drag y velocidad del fluido (V_fluido) ya definidos
Eficiencia = 0.85  # Asumiendo una eficiencia del 85%

# Process
P_util = F_drag * V_fluido
P_mec = P_util / Eficiencia

# Output
print(f"Potencia mecánica requerida: {P_mec} W")
```

## 5. Energía en carrera

```python
# Input
# P_mec ya definido
Distancia = x "(m)"

#Process
tiempo_viaje = Distancia / V_fluido
E_tot = P_mec * tiempo_viaje

# Output
print(f"Energía total requerida para el viaje: {E_tot} J")
```
