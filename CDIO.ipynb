## Objetivo específico

Transportar el vehículo de aproximadamente 2.5 kg desde un punto A hacia un punto B de la pista de competencia que posee un total de 64 metros  en el menor tiempo posible. Además, el diseño está sujeto a las siguientes restricciones:

- Peso máximo de operación: 2.5 kg
- Distancia total pista de carrera: 64m.
- Curva más cerrada: Radio 1,25m.
- La valoración de los elementos/materiales adicionales no puede superar los 15.000 $ en total.
- La valoración de los elementos usados/donados corresponde al 50% del precio nuevo comercial puesto en Chile.
- Los materiales/elementos no pueden superar el 15% del peso bruto del auto.
